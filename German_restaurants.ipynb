```python
# Since mode of the price category in German restaurants is "€20–30", we can fill missing values with the mode of the price category.
```

```python
x = "€20–30"
grc["price"].fillna(x)
```

```python
grc.info()
```

```python
grc["price"].value_counts()
```

```python
# In addition to missing values, there are also incorrect values such as "$93" or "$$". We can assume that dollar sign or euro sign data
# represent price range. We have some specific price ranges such as €10–20 or €20–30. In our data, "€10–20" is "€" or "$", "€20–30" is "€€" or "$$" etc.
# Also, we can include price values ​​that are not in range($93). If we limit our range to "€50+", values ​​over 50€ will be included in "€50+".
```

```python
grc_backup = grc.copy()
```

```python
grc_backup
```

```python
for x in grc.index:
    if grc.loc[x, "price"] in ["$", "£"]:
        grc.loc[x, "price"] = "€10–20"
    elif grc.loc[x, "price"] in ["$$", "££"]:
        grc.loc[x, "price"] = "€20–30"
    elif grc.loc[x, "price"] in ["$$$", "£££"]:
        grc.loc[x, "price"] = "€30–40"
    elif grc.loc[x, "price"] in ["€100+", "$93", "$98", "£82", "£88", "$51"]:
        grc.loc[x, "price"] = "€50+"
    elif grc.loc[x, "price"] == "€10–30":
        grc.loc[x, "price"] = "€20–30"
    elif grc.loc[x, "price"] == "€20–40":
        grc.loc[x, "price"] = "€20–30"
```

```python
grc["price"].value_counts()
```

```python
grc.to_csv("german_restaurants_cleaned.csv", index=False)
```
